```python
import numpy as np
from scipy.stats import multivariate_normal
import plotly.graph_objects as go

mu = np.array([0, 0]) 
Sigma = np.array([[1, 0.8], [0.8, 1]]) 

x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

rv = multivariate_normal(mu, Sigma)
Z = rv.pdf(pos)

fig = go.Figure(data=[
    go.Surface(
        x=X, y=Y, z=Z,
        colorscale='Viridis',
        hovertemplate=(
            'X: %{x:.2f}<br>'
            'Y: %{y:.2f}<br>'
            'probability density: %{z:.4f}<extra></extra>'
        )
    )
])

fig.update_layout(
    title='N(mu, SIGMA)',
    scene=dict(
        xaxis_title='X axis',
        yaxis_title='Y axis',
        zaxis_title='probability density',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=0.6) 
        )
    ),
    width=800,
    height=600,
    autosize=False,
    margin=dict(l=65, r=50, b=65, t=90)
)


fig.show()
```
