```python
#ball
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
import plotly.graph_objects as go

np.random.seed(100)

def generate_3d_normal(n=500, cov_matrix=np.eye(3)):
    u1 = np.random.uniform(size=n)
    u2 = np.random.uniform(size=n)
    u3 = np.random.uniform(size=n)
    u4 = np.random.uniform(size=n)
    r1 = np.sqrt(-2 * np.log(u1))
    theta1 = 2 * np.pi * u2
    z0 = r1 * np.cos(theta1)
    z1 = r1 * np.sin(theta1)
    r2 = np.sqrt(-2 * np.log(u3))
    theta2 = 2 * np.pi * u4
    z2 = r2 * np.cos(theta2)
    points = np.vstack([z0, z1, z2]).T
    if not np.array_equal(cov_matrix, np.eye(3)):
        eigvals, eigvecs = np.linalg.eig(cov_matrix)
        sqrt_cov = eigvecs @ np.diag(np.sqrt(eigvals)) @ np.linalg.inv(eigvecs)
        points = points @ sqrt_cov
    return points[:, 0], points[:, 1], points[:, 2]
x, y, z = generate_3d_normal(n=1000)
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=3,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(
    title='3D Normal Distribution',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    width=800,
    height=600
)

fig.write_html("3d_normal_rotation.html")
fig.show()
```

```python
#cigar
import numpy as np
import plotly.graph_objects as go

np.random.seed(100)

def generate_3d_normal(n=500, cov_matrix=np.eye(3)):
    u1 = np.random.uniform(size=n)
    u2 = np.random.uniform(size=n)
    u3 = np.random.uniform(size=n)
    u4 = np.random.uniform(size=n)
    r1 = np.sqrt(-2 * np.log(u1))
    theta1 = 2 * np.pi * u2
    z0 = r1 * np.cos(theta1)
    z1 = r1 * np.sin(theta1)
    r2 = np.sqrt(-2 * np.log(u3))
    theta2 = 2 * np.pi * u4
    z2 = r2 * np.cos(theta2)
    points = np.vstack([z0, z1, z2]).T
    if not np.array_equal(cov_matrix, np.eye(3)):
        eigvals, eigvecs = np.linalg.eig(cov_matrix)
        sqrt_cov = eigvecs @ np.diag(np.sqrt(eigvals)) @ np.linalg.inv(eigvecs)
        points = points @ sqrt_cov
    return points[:, 0], points[:, 1], points[:, 2]


cov_matrix = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 10]
])

x, y, z = generate_3d_normal(n=1000, cov_matrix=cov_matrix)

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=3,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(
    title='3D Normal Distribution (Cigar-shaped)',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    width=800,
    height=600
)

fig.show()
fig.write_html("Cigar_3d_normal_rotation.html")
```
