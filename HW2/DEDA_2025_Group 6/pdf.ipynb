```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_distribution(mean, std_dev):
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
    y = norm.pdf(x, mean, std_dev)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', linewidth=2, label=f'μ = {mean}, σ = {std_dev}')
    plt.title('Normal (Gaussian) Distribution', fontsize=15)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)
    plt.show()

if __name__ == "__main__":
    try:
        mean = float(input("Enter mean (μ): "))
        std_dev = float(input("Enter standard deviation (σ): "))
        if std_dev <= 0:
            print("Standard deviation must be positive!")
        else:
            plot_normal_distribution(mean, std_dev)
    except ValueError:
        print("Please enter valid numbers!")
```
